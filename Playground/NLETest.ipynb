```python
env.reset()
for i in range(500):
    action = random.sample(range(16), k=1)[0]
    try:
        observation, reward, done, info = env.step(action)
    except IndexError:
        print(action)
        break
    if i % 50 == 0:
        env.render()
        print(f'Reward: {reward}, Done: {done}, Info: {info}')
    if done:
        print(f'Total {i + 1} timesteps.')
        break    
```
