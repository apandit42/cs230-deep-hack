```python
# Reduced action space
ALLOWED_ACTIONS = tuple(
    # Compass directions (N,E,S,W)
    list(nethack.CompassDirection)
    # Go up stairs, down stairs, wait
    + list(nethack.MiscDirection)
    # All Character Actions
    + [
        Command.APPLY,
        Command.CAST,
        Command.CLOSE,
        Command.DIP,
        Command.DROP,
        Command.EAT,
        Command.ESC,
        Command.FIRE,
        Command.FORCE,
        Command.INVOKE,
        Command.KICK,
        Command.LOOT,
        Command.OFFER,
        Command.OPEN,
        Command.PAY,
        Command.PICKUP,
        Command.PRAY,
        Command.PUTON,
        Command.QUAFF,
        Command.QUIVER,
        Command.READ,
        Command.REMOVE,
        Command.RIDE,
        Command.RUB,
        Command.SEARCH,
        Command.TAKEOFF,
        Command.THROW,
        Command.TIP,
        Command.UNTRAP,
        Command.WEAR,
        Command.WIELD,
        Command.WIPE,
        Command.ZAP,
        nethack.TextCharacters.SPACE
    ]
    # All Raw Key Presses
    + [
        ord(x) for x in list(string.ascii_letters)
    ]
)
```
